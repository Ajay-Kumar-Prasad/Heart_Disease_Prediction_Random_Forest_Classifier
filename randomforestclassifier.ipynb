```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('heart.csv')
data
```

```python
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train,Y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_pred,Y_test)
print("Accuracy:", score)
```

```python
print(classification_report(y_pred,Y_test))
```

### #With Hyperparameter tuning# 

```python
#parameters
n_estimators = [20,60,100,120]
max_features = [0.2,0.6,1.0]
max_depth = [2,8,None]
max_samples = [0.5,0.75,1.0]

param_grid = {
    'n_estimators':n_estimators,
    'max_features':max_features,
    'max_depth':max_depth,
    'max_samples':max_samples
}
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier2 = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(estimator = classifier2, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)
```

```python
cv.fit(X_train,Y_train)
```

```python
cv.best_params_
```

```python
cv.best_score_
```

### For Large Dataset Use RandomSearchCV ### 

```python
bootstrap = [True,False]
min_samples_split = [2,5]
min_samples_leaf = [1,2]
```

```python
param_grid2 = {
    'n_estimators':n_estimators,
    'max_features':max_features,
    'max_depth':max_depth,
    'max_samples':max_samples,
    'bootstrap':bootstrap,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rscv = RandomizedSearchCV(estimator = classifier2, param_distributions=param_grid2, cv=5, verbose=2, n_jobs=-1)
```

```python
rscv.fit(X_train,Y_train)
```

```python
rscv.best_params_
```

```python
rscv.best_score_
```
